```python
def calcDistFromAns(matrix, n):
  total = 0
  for i in range(n):
    for j in range(n):
      if(matrix[i][j] == 0):
        x = n-1
        y = n-1
      else:
        x = int(matrix[i][j]/n)
        y = matrix[i][j]%n - 1
        if(y<0):
          y+=n
          x-=1
      diff = abs(x - i) + abs(y - j)
      total += diff
  return total
```

```python
import random

def generateRandomMatrix(n = 5):
  dic = dict()
  nums = []

  for i in range(1,n*n):
    nums.append(i)

  matrix = []

  for i in range(n):
    temp = []
    for j in range(n):
      if(len(nums) == 0):
        temp.append(0)
        break
      index = random.randint(0, len(nums)-1)
      temp.append(nums[index])
      nums.pop(index)
    matrix.append(temp)
  dic[0] = matrix
  dic[1] = [n-1,n-1]
  dic[2] = calcDistFromAns(matrix, n)
  dic[3] = []
  dic[4] = 0
  if(not IsSolvable(dic)):
    temp = dic[0][n-1][n-3]
    dic[0][n-1][n-3] = dic[0][n-1][n-2]
    dic[0][n-1][n-2] = temp
  return dic
```

```python
def checkInside(n, i, j):
  if(i<0 or i>=n or j <0 or j>=n):
    return False
  return True
```

```python
def isAnswer(matrix, n):
  start = 0
  for i in range(n):
    for j in range(n):
      if(matrix[i][j] != start):
        return False
      start += 1
  return True
```

```python
def compare2DArrays(arr1, arr2):
  for i in range(len(arr1)):
    for j in range(len(arr1[0])):
      if(arr1[i][j] != arr2[i][j]):
        return False
  return True
```

```python
def addNode(q, newNode):
  leastIndex = 0
  start = 0
  end = len(q) - 1
  # print(newNode[2], newNode[3])
  # reqNum = newNode[2] + newNode[4]
  reqNum = newNode[2]
  index = int((start+end)/2)
  while(end - start >= 0):
    # print(start, index, end)
    index = int((start+end)/2)
    qstrt = q[start][2]
    qend = q[end][2]
    if(end-start ==0):
      if(qstrt < reqNum):
        q.append(newNode)
        return
      elif(qstrt > reqNum):
        q.insert(start, newNode)
        return
    if(end - start == 1):
      if(qstrt >= reqNum):
        q.insert(start, newNode)
        return
      if(end >= reqNum):
        q.insert(start+1, newNode)
        return
      if(end < reqNum):
        q.insert(start + 2, newNode)
        return
    midValue = q[index][2]
    if(midValue > reqNum):
      end = index
    elif(midValue < reqNum):
      start = index
    else:
      break
  q.insert(index, newNode)
```

```python
def getInvCount(matrix):
    N = len(matrix)
    M = len(matrix[0])
    arr1=[]
    for y in matrix:
        for x in y:
            arr1.append(x)
    arr=arr1
    inv_count = 0
    for i in range(N * M - 1):
        for j in range(i + 1,N * M):
            if (arr[j] and arr[i] and arr[i] > arr[j]):
                inv_count+=1
    return inv_count
```

```python
def IsSolvable(item):
  matrix = item[0]
  blank = item[1]
  i = blank[0]
  j = blank[1]
  no_of_inv = getInvCount(matrix)
  if(len(matrix) % 2 == 1):
    if(no_of_inv % 2 == 0):
      return True
  else:
    if((len(matrix) - i)%2 == 0):
      if(no_of_inv%2 == 1):
        return True
    else:
      if(no_of_inv%2 == 0):
        return True
  return False
```
